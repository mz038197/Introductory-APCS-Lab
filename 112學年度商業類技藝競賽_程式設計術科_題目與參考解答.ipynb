## Problem F 質因數分解

### 題目（PDF 擷取）

```text
Problem Description
自然數是數學中的一個基本概念。自然數的定義通常包含從1 開始的所有正整數（1, 2, 3, 4,
5, ...）。1 不是質數。根據質數的定義，一個質數必須是一個大於1 的自然數，且除了1 和
它本身之外，不能被其他自然數整除。1 只有一個因數（它本身），所以它不符合質數的定
義。因此，在數學中，1 被分類為非質數、非合數的獨特數字。
在進行質因數分解時，通常不包括1。質因數分解是將一個數字分解為它的質數因子的乘
積。
質因數分解就是把一個數字，切分為數個質數的乘積，例如12=2*2*3。
輸出質因數分解，用空格隔開質因數，先不處理次方、乘號，如12=2 2 3。
Input Format
每組輸入只有一列，包含一個整數𝑁，2 ≤𝑁≤106。
Output Format
針對輸入整數輸出一個質因數分解字串，用空格隔開質因數，由小到大。
Sample Input 1
20
Sample Output 1
2 2 5
Sample Input 2
17
Sample Output 2
17
Sample Input 3
999997
Sample Output 3
757 1321
Hint
如何如何判斷是否為因數？
在程式中要判斷𝑎是否為𝑏的因數可以透過除法取餘數是否為零來判斷：
例如𝑛mod 2 = 0 即代表𝑛是2的倍數翻成中文是「𝑛除以2，餘數為0」、「𝑛被2 整除」、
「2 整除𝑛」
```

### 參考解答


````python
N = int(input())

def is_prime(n):
  if n <= 1:
        return False
  if n == 2:
      return True

  if n % 2 == 0:
      return False

  for i in range(3, int(n**0.5) + 1, 2):
      if n % i == 0:
          return False
  return True

def recursive(n, arr):
  for i in range(2, n+1):
    if n % i == 0 and is_prime(i):
      arr.append(i)
      recursive(n // i, arr)
      break

arr = []
recursive(N, arr)
print(*arr)
````

## Problem H 數字運算(減1或除以10)

### 題目（PDF 擷取）

```text
Problem Description
數字運算(減1或除以10)
如果數字的最後1位不是零，就將數字減1；如果數字的最後1位是零，就將數字除以10（即
去掉最後一位）。
給定一個整數𝑛。對這個數字進行𝑘次數字運算。你的任務是在進行𝑘次減1或除以10的運算
後，輸出最終結果。保證最終結果會是正整數。正整數大於0的整數。
Input Format
每組輸入只有一列，包含兩個整數𝑛和𝑘, 2 ≤𝑛≤109, 1 ≤𝑘≤50）分別是要數字運算和
減1或除以10的次數。
Output Format
輸出一個整數，將𝑛減1或除以10 𝑘次的結果。保證結果將是正整數。
Sample Input 1
512 4
Sample Output 1
50
Sample Input 2
1000000000 9
Sample Output 2
1
Sample Input 3
1001 2
Sample Output 3
100
Hint
解題思路：
在Sample Input 1中，對應於以下序列：512 →511 →510 →51 →50。
```

### 參考解答


````python
N, K = map(int, input().split())

def recursive(last, curr, times):

  if times == K:
    return curr

  if curr <= 0:
    return last

  times += 1
  last = curr

  if str(curr)[-1] != "0":
    curr -= 1
  else:
    curr //= 10

  return recursive(last, curr, times)

print(recursive(N, N, 0))
````


### 參考解答


**（註：原檔標為會超時的解法）**

````python
#會超時的解法
times = int(input())
for _ in range(times):
  result = set()
  n = int(input())
  s = list(input())
  for i in range(n-1):
    d = "".join(s[:i] + s[i+2:])
    result.add(d)
  print(len(result))
````

**（註：原檔標為較易讀的解法）**

````python
#好懂得解法
times = int(input())

results = []
for _ in range(times):
    n = int(input())
    s = list(input())

    # 總共有 n-1 種刪除連續兩個字元的方法
    ans = n - 1

    # 若 s[i-1] == s[i+1]，會導致重複結果
    for i in range(1, n - 1):
        if s[i-1] == s[i+1]:
            ans -= 1

    results.append(str(ans))

print("\n".join(results))
````


### 參考解答


````python
class Graph:
  def __init__(self):
    self.adjacentlist = {}

  def addVertex(self, node):
    self.adjacentlist[node] = []

  def addEdge(self, node1, node2):
    self.adjacentlist[node1].append(node2)
    self.adjacentlist[node2].append(node1)

  def bfs(self, start):

    if start not in self.adjacentlist:
      return

    dist = {node:-1 for node in self.adjacentlist.keys()}
    dist[start] = 0
    far = start

    visited = []
    queue = [start]
    while queue:
      node = queue.pop(0)
      if node not in visited:
        visited.append(node)
        for neighbor in self.adjacentlist[node]:
          if neighbor not in visited:
            dist[neighbor] = dist[node] + 1
            if dist[neighbor] > dist[far]:
              far = neighbor
            queue.append(neighbor)

    return dist, far

numNode = int(input())
graph = Graph()

for i in range(1, numNode+1):
  graph.addVertex(i)

for i in range(numNode-1):
  node1, node2 = map(int, input().split())
  graph.addEdge(node1, node2)

_, A = graph.bfs(1) #先找出任一點(ex:1)最長距離的節點
distA, B = graph.bfs(A) #再用此節點找出最長距離的節點
print(distA[B])
````

```text
component）內所有點的鄰居都等於2，是就成循環圖(cycle graph)，不是就非循環圖
```

### 參考解答


````python
class Graph:
  def __init__(self):
    self.adjacentlist = {}

  def addVertex(self, node):
    self.adjacentlist[node] = []

  def addEdge(self, node1, node2):
    self.adjacentlist[node1].append(node2)
    self.adjacentlist[node2].append(node1)

  def cycle_count(self):

    visited = []
    circle_count = 0

    #選一個起點
    for start in self.adjacentlist.keys():

      if start in visited or len(self.adjacentlist[start]) != 2:
        continue

      queue = []
      is_finded = True
      queue.append(start)

      while queue:

        node = queue.pop(0)

        if node not in visited:
          visited.append(node)

          for neighbor in self.adjacentlist[node]:

            #檢查節點是否2連接
            if len(self.adjacentlist[neighbor]) != 2:
              is_finded = False

            #檢查是否拜訪過
            if neighbor not in visited:
              queue.append(neighbor)

      if is_finded:
        circle_count += 1

    return circle_count


numNode, numEdge = map(int, input().split())
graph = Graph()

for i in range(1, numNode+1):
  graph.addVertex(i)

for i in range(numEdge):
  node1, node2 = map(int, input().split())
  graph.addEdge(node1, node2)

count = graph.cycle_count()
print(count)
````
