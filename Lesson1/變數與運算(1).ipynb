## **重點複習**

*   變數可以儲存數值等各式資料。
*   如果有需要互相引用的變數，在使用前要先定義該變數的內容。
*   #start 是一個註解，用於維持程式結構。
*   資料型態有助於區別變數的類型。
*   Python 有 +, -, *, /, % 五種算術運算子，另外還有 **, // 兩種特殊的運算子。
*   str() 能將數值型態的資料轉換為文字。
