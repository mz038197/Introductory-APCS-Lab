### 題目（PDF 擷取）

```text
Problem Description
密碼通常包含大小寫字母、數字和特殊字符。給定一個密碼（字串），請撰寫一程式，每項
合格得1 分，總分為5 分，檢查密碼規則如下：
1. 密碼長度至少要有八個字元(含)以上；
2. 至少要有一個數字；
3. 至少要有一個大寫英文字母；

> ]
Input Format
一個密碼字符串（長度不超過100）。
Output Format

Sample Input 1

Sample Input 2

Sample Output 5

Sample Input 6
123456
Sample Output 6
1
Sample Input 7
@@@ ###
Sample Output 7

```

### 參考解答


````python
class PasswordValidator:

  def __init__(self):
    self.score = 0

  def validate(self, password):
    if self._checkLength(password):
      self.score += 1
    if self._checkNumerical(password):
      self.score += 1
    if self._checkUpper(password):
      self.score += 1
    if self._checkLower(password):
      self.score += 1
    if self._checkSpecial(password):
      self.score += 1

  def _checkLength(self, password):
    return len(password) >= 8

  def _checkNumerical(self, password):
    for ch in password:
      if ch.isnumeric():
        return True
    return False

  def _checkUpper(self, password):
    for ch in password:
      if ch.isupper():
        return True
    return False

  def _checkLower(self, password):
    for ch in password:
      if ch.islower():
        return True
    return False

  def _checkSpecial(self, password):
    special = [
        '[', '!', '@', '#', '$', 'ˆ', '&',
        '*', '(', ')', '?', '”', ':', '{',
        '}', '|','<', '>', ']'
        ]

    for ch in password:
      if ch in special:
        return True
    return False

password = list(input())
validator = PasswordValidator()
validator.validate(password)
print(validator.score)
````

````python
def get_order(num, orders, result):
  if len(orders) == 3:
    s = sum([x**2 for x in orders])
    if s == num:
      result.append(orders.copy())
    return

  for i in range(0, num+1):
    if len(orders) != 0 and i < orders[-1]:
      continue

    orders.append(i)
    get_order(num, orders, result)
    orders.pop()

orders = []
result = []

num = int(input())
get_order(num, orders, result)

if len(result) == 0:
  print(-1)
else:
  print(result[0])
````

## Problem N 運算式轉換

### 題目（PDF 擷取）

```text
Problem Description
運算子（operator）通俗的意思是「符號」，像是+ - * / 之類的；而運算元（operand）意
思是「數字」或「代數」，像是1, 2, 3, . . . 或𝑥, 𝑦, 𝑧, . . . 等都是運算元。
一般我們常看到的運算式，像是𝑎+𝑏×𝑐或𝑎/𝑏−𝑐都是中序運算式(Infix Notation)，可是電腦不
懂得如何處理中序運算式，所以電腦會先把中序運算式轉成後序運算式(Postfix Notation)或
前序運算式(Prefix Notation)再計算。
想以人工轉換計算後序式的話，可以使用括號法，將運算子兩旁的運算元，依先後順序全部
括號起來，然後將右括號取代為左邊最接近的運算子（從最內層括號開始），最後去掉全部
的左括號就可以完成。例如：
𝑎+ 𝑏* 𝑑+ 𝑐/𝑑
((𝑎+ (𝑏* 𝑑)) + (𝑐/𝑑))
𝑎𝑏𝑑* +𝑐𝑑/+
另一個方式是堆疊法，使用迴圈取出中序式的元素，遇運算元直接輸出，遇到運算子與左括
號進行堆疊，堆疊中運算子優先順序若大於等於讀入的運算子優先順序，直接輸出堆疊中的
運算子，再將讀入的運算子置入堆疊；遇右括號輸出堆疊中的運算子至左括號。
Input Format
每筆測資一行，每一行會有一個中序運算式，每個運算子及運算元都會以空格分隔。
運算元有6種’+’, ’-’, ’*’, ’/’, ’(’, ’)’，分別代表加減乘除與括號。
優先順序: 括號>乘除>加減，運算子是小寫或大寫的英文字母(代數)或數字。
Output Format
輸出順序由左到右轉換好的後序運算式，每個運算子及運算元以空格分隔。
Sample Input 1
a + b * c
Sample Output 1
a b c * +
Sample Input 2
a / b - c
Sample Output 2
a b / c -
Sample Input 3
a + b * ( c * ( d + e ) )
Sample Output 3
a b c d e + * * +
Sample Input 4
( ( a + b ) - ( c + d ) ) * d
Sample Output 4
a b + c d + - d *
14
```

### 參考解答


````python
ops = {"+":1, "-":1, "*":2, "/":2, "(":3, ")":3}

s = input().split()
stack = []

result = ""
for data in s:
  if data not in ops:
    result += data
  elif data == ")":
    while stack[-1] != "(":
      result += stack.pop()
    stack.pop()
  elif (len(stack) == 0) or (stack[-1] == "(") or (ops[stack[-1]] < ops[data]) :
    stack.append(data)
  else:
    result += stack.pop()
    stack.append(data)

while len(stack) != 0:
  result += stack.pop()

print(" ".join(result))
````

## Problem O 後序運算式求值

### 題目（PDF 擷取）

```text
Problem Description
後序運算式，又稱逆波蘭表示法，是一種運算式的記法。平常我們使用的運算式如𝐴+ 𝐵為
中序運算式，即運算子在兩運算元中間。可是這種表示法必須用括弧來表示運算順序，電
腦不易解析。因此，現代電腦大多利用堆疊與後序運算式以減少記憶體用量。後序運算式

Input Format
輸入只有一行，含有一個後序運算式。
運算元為32 位元有號整數，運算子包括+, -, *, 及/。
所有的運算元及與運算子之間均以一個空白隔開。運算過程及結果也都是32 位元有號整數。
為避免小數誤差問題，運算結果必定為整數，運算過程中也必定不會出現小數的結果，因此

Output Format
輸出該運算式所求得的值。
Sample Input 1
1 2 3 * +
Sample Output 1
7

9
15
```

### 參考解答


````python
s = input().split()

stack = []
for i in range(len(s)):
  if s[i] not in ["+", "-", "*", "/"]:
    stack.append(int(s[i]))
  else:
    num2 = stack.pop()
    num1 = stack.pop()
    if s[i] == "+":
      stack.append(num1 + num2)
    elif s[i] == "-":
      stack.append(num1 - num2)
    elif s[i] == "*":
      stack.append(num1 * num2)
    else:
      stack.append(num1 // num2)

print(stack[0])
````

## Problem P 不同值的查詢

### 題目（PDF 擷取）

```text
Problem Description
給你一個有𝑛個元素的陣列，以及𝑞筆查詢，每筆查詢求[𝑎, 𝑏] 有多少不同元素。
若測資很大的時候，莫隊演算法(Mo’s Algorithm)是一種用來處理區間問題的離線算法，也
就是說，它會在把所有詢問讀進來之後，再利用詢問之間的關聯性來降低運算時間。
要使用莫隊演算法有一個先決條件：如果已經知道[𝑙, 𝑟]的答案，那必須要可以快速地知
道[𝑙, 𝑟+ 1] [𝑙, 𝑟−1] [𝑙+ 1, 𝑟] [𝑙−1, 𝑟] 的答案，也就是把左或右界移動一個位置，算出新答案
的。
Input Format
第一行有兩個整數𝑛與𝑞，𝑛(1 ≤𝑛, 𝑞≤105)，代表陣列長度以及查詢數量。
接下來有𝑛個整數𝑥1, 𝑥2, . . . , 𝑥𝑛，(1 ≤𝑥𝑖≤106)，代表陣列的元素。
最後，有𝑞行查詢，每行有兩個整數𝑎與𝑏，(1 ≤𝑎< 𝑏≤𝑛)。每兩個數之間用一個空格隔

```

### 參考解答


````python

temp = {}

def add(value):
  if value not in temp:
    temp[value] = 1
  else:
    temp[value] += 1

def delete(value):
  if value in temp:
    temp[value] -= 1
    if temp[value] <= 0:
      del temp[value]

n, numQuery = map(int, input().split())
arr = list(map(int, input().split()))

result = [0]*numQuery
queries = []

for index in range(numQuery):
  query = input().split()
  query = [int(qua)-1 for qua in query]
  query.append(index)
  queries.append({"left":query[0], "right":query[1], "index":index})

queries = sorted(queries, key=lambda query: (query["left"], query["right"] if query["left"]%2 == 0 else -query["right"]))

left = 0
right = -1

for query in queries:
  while left > query["left"]:
    left -= 1
    add(arr[left])

  while right < query["right"]:
    right += 1
    add(arr[right])

  while left < query["left"]:
    delete(arr[left])
    left += 1

  while right > query["right"]:
    delete(arr[right])
    right -= 1

  result[query["index"]] = len(temp)

print(*result)

````

## Problem Q 樹的距離

### 題目（PDF 擷取）

```text
Problem Description
這是一棵由𝑛個節點組成的樹。找出每個節點到其他節點的最大距離。
Figure 1: Sample Input
Input Format
第一行輸入包含一個整數𝑛：節點的數量。節點編號為1, 2, . . . , 𝑛。
接著有𝑛−1 行描述邊。每行包含兩個整數𝑎和𝑏：節點𝑎與𝑏之間存在一條邊（以空格隔
開），1 ≤𝑛≤106, 1 ≤𝑎, 𝑏≤𝑛。保證輸入是一棵樹。
Output Format
輸出𝑛個整數: 所有節點1, 2, . . . , 𝑛到其他點的最大距離（以空格隔開）。
Sample Input 1
5
1 2
1 3
3 4
3 5
Sample Output 1
2 3 2 3 3
Sample Input 2
7
1 2
1 3
1 4
1 5
1 6
1 7
Sample Output 2
1 2 2 2 2 2 2
```

### 參考解答


````python
class Graph:
  def __init__(self):
    self.adjacentlist = {}
    self.nodeCount = 0

  def addVertex(self, node):
    if node not in self.adjacentlist:
      self.adjacentlist[node] = []
      self.nodeCount += 1

  def addEdge(self, node1, node2):
    self.adjacentlist[node1].append(node2)
    self.adjacentlist[node2].append(node1)

  def bfs(self, start):

    if start not in self.adjacentlist:
      return

    dist = {node:-1 for node in self.adjacentlist.keys()}
    dist[start] = 0
    far = start

    visited = []
    queue = [start]
    while queue:
      node = queue.pop(0)
      if node not in visited:
        visited.append(node)
        for neighbor in self.adjacentlist[node]:
          if neighbor not in visited:
            dist[neighbor] = dist[node] + 1
            if dist[neighbor] > dist[far]:
              far = neighbor
            queue.append(neighbor)

    return dist, far


numNode = int(input())
graph = Graph()

for i in range(1, numNode+1):
  graph.addVertex(i)

for i in range(numNode-1):
  node1, node2 = map(int, input().split())
  graph.addEdge(node1, node2)

_, A = graph.bfs(1)
distA, B = graph.bfs(A)
distB, _ = graph.bfs(B)

result = [max(distA[i], distB[i]) for i in range(1, numNode + 1)]
print(" ".join(list(map(str, result))))
````
