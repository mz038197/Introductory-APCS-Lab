#### **計算時間**
#### 將程式碼替換為 Python 的內建排序，就能透過執行時間比較兩者的效率了

````python
#start
import random
import time
data = list(range(1, 1000))
random.shuffle(data)

t1 = time.time()
data.sort()
t2 = time.time()

print("執行時間:", t2 - t1, "s")
````

````python
#start
import random
import time

def bubble_sort(array):
    sorted_list = array.copy()
    N = len(sorted_list)
    for i in range(N-1):
        for j in range(N - i - 1):
            if sorted_list[j] > sorted_list[j + 1]:
                sorted_list[j], sorted_list[j + 1] = sorted_list[j + 1], sorted_list[j]
    return sorted_list

data = list(range(1, 1000))
random.shuffle(data)

t1 = time.time()
a = bubble_sort(data)
t2 = time.time()

print("執行時間:", t2 - t1, "s")
````

### **認識 worst-case**
*   代表這個演算法效能最差的情況 (事倍功半)

#### 以氣泡排序法為例:
*   worst-case 就是這組數列在排序之前的狀況是由大到小
*   需要最多的執行次數，耗費更多的計算時間

#### best-case
```python
data = list(range(1, 1000))
```

#### avg-case
```python
data = list(range(1, 1000))
random.shuffle(data)
```

#### worst-case
```python
data = list(range(1, 1000))
data.reverse()
```

### **2. 搜尋與極值**

*   在一串連續的資料當中，我們可能會想要這串資料中的某個數值

#### **極值 (MAX, min)**
