#### **泡泡排序法**
*   傳入指定參數，詢問回傳值
*   逐步推算即可得到計算結果



#### **例題**
#### 透過觀看 [泡泡排序法](https://www.youtube.com/watch?v=nmhjrI-aW5o) 的演算法影片，學習製作一個 bubble sort 功能

#### 程式實作分為三個部分：
*   亂數隨機產生數列
*   泡泡排序演算法
*   輸出排序後的結果




````python
def bubble_sort(array):
    sorted_list = array.copy()
    N = len(sorted_list)
    for i in range(N-1):
        for j in range(N - i - 1):
            if sorted_list[j] > sorted_list[j + 1]:
                sorted_list[j], sorted_list[j + 1] = sorted_list[j + 1], sorted_list[j]
    return sorted_list
````

#### **計算時間**
#### 將程式碼替換為 Python 的內建排序，就能透過執行時間比較兩者的效率了

````python
import random
import time
data = list(range(1, 1000))
random.shuffle(data)

t1 = time.time()
data.sort()
t2 = time.time()

print("執行時間:", t2 - t1, "s")
````

### **搜尋與極值**

#### **搜尋 (search)**
*   在一串連續的資料當中，我們可能會想要這串資料中的某個數值

#### **極值 (MAX, min)**
*   在一串連續的資料當中，資料的最大值和最小值，也是相當有意義

#### **常遇問題**
*   設計不良則 worst-case 會無法通過，以 zerojudge 解題來說很有可能會得到 TLE 的結果

#### **搜尋種類**
*   線性搜尋 : 一個一個尋找，從頭掃到尾
*   二分搜尋 : 資料先經過排序，使用二分搜尋法，效率會遠比線性搜尋好很多



#### **搜尋例題**
#### 電腦會隨機產生一個數字，範圍在 1~100 之間。
#### 請你使用「線性搜尋」和「二分搜尋」的方式，找出電腦的產生的數字，並記錄搜尋錯誤的次數。

##### **線性搜尋**

````python
def search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return i
  return -1

a = [1, 2, 4, 6, 8, 9]
print(search(a, 9))
````

##### **二分搜尋**

````python
# Iterative Binary Search Algorithm
def binary_search(data, target):
  left = 0
  right = len(data) - 1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return mid
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1

a = [1, 2, 4, 6, 8, 9]
print(binary_search(a, 10))
````

````python
# Recursive Binary Search Algorithm
def binary_search(data, low, high, target):

  if low > high:
    return -1

  mid = (low + high) // 2
  if data[mid] == target:
    return mid

  if data[mid] < target:
    return binary_search(data, mid + 1, high, target)
  else:
    return binary_search(data, low, mid-1, target)

a = [1, 2, 4, 6, 8, 9]
print(binary_search(a, 0, len(a)- 1, 9))
````

#### **極值例題**
#### 輸入一個 N 值，使用亂數隨機產生 N 個數字，範圍在1~100之間
#### 請找到串列中最大和最小的值。

````python
a = list(range(1, 100))
sorted_list = sorted(a)
print(sorted_list)
print("Min value: "+ str(sorted_list[0]))
print("Max value: "+ str(sorted_list[-1]))
````

## **重點整理**
*   搜尋方法基礎的有兩種，「線性搜尋法」與「二分搜尋法」
*   尋找極值的方式，可以先將目標的資料群排序後，頭尾即是最大和最小值
*   搜尋法、尋找極值都與 worst-case 有關，如果使用錯誤的方法，雖然程式可以找到結果，但是卻會嚴重影響程式效能
