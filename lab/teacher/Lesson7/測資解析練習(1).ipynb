````python
### BEGIN SOLUTION

import sys

for s in sys.stdin:
    nums = list(map(int, s.split()))
    N = nums[0]

    for i in range(1, N+1):
        print(nums[i], end=' ')
    print()
    
### END SOLUTION
````

#### **10-2a 題目說明：**
#### 撲克牌遊戲，輸入 N 張撲克牌花色與數字，輸出最大的牌面
#### 輸入說明：

*   每組測資一共有兩行。
*   第一行為一個數字，代表 N 張牌。
*   第二行為英文與數字交錯代表花色與點數。
*   S(黑桃)、H(紅心)、D(方塊)、C(梅花)。

```python
Sample Input:
5
S 1 H 9 D 9 H 13 S 12
3
H 3 C 10 S 8
```

````python
### BEGIN SOLUTION
while True:
    try:
        #start
        N = int(input())
        s = input()
        data = s.split()

        stack = []
        i=0
        while i<2*N:
            card = data[i]+data[i+1]
            #print(card)
            stack.append(card)
            i += 2
        print(stack)

    except:
        break
### END SOLUTION
````

#### 畢業典禮 [e788](https://zerojudge.tw/ShowProblem?problemid=e788)，輸入一組學號與學生姓名，輸出排序的結果
#### **輸入說明：**

*   每組測資有 N+1 行。
*   第一行為一個數字，代表有 N 個人要排序。

```python
Sample Input:
5
```

````python
### BEGIN SOLUTION
while True:
    try:
        #start
        N = int(input())

        student = []
        for i in range(N):
            s = input()
            si = [ x for x in s.split()]
            student.append(si)

        print(student)


    except:
        break
### END SOLUTION
````

#### **10-2c 題目說明：**
#### 輸入一個運算式，其中分數與運算符號皆由空格隔開，[e941](https://zerojudge.tw/ShowProblem?problemid=e941)
#### **輸入說明：**

*   分數格式皆為 X/Y，XY為正整數。
*   出現的分數最多五個。

```python
Sample Input #1:   Sample Output #1
1/2 + 1/3 - 1/6    2/3
```
```python
Sample Input #2:   Sample Output #2
1/3 - 1/2          -1/6
```

````python
### BEGIN SOLUTION

s = input()
data = [ x for x in s.split()]

N = len(data)

num = []

for i in range(0, N):
    temp = list(data[i])
    num.append(temp)

print(num)

### END SOLUTION
````

#### **試試完成完整版本!**

````python
### BEGIN SOLUTION

def gcd(a, b):
  c = max(a,b) % min(a, b)
  if c == 0:
    return min(a, b)
  return gcd(min(a, b), c)

def normalize(data):
    n, d = data
    
    if n == 0:
        return [0, 1]

    abs_data = [abs(x) for x in data]
    g = gcd(abs_data[0], abs_data[1])
    return [n // g, d // g]

def plus(f1, f2):
  n = f1[0]*f2[1]+f1[1]*f2[0]
  d = f1[1]*f2[1]
  return normalize([n, d])

def minus(f1, f2):
  n = f1[0]*f2[1]-f1[1]*f2[0]
  d = f1[1]*f2[1]
  return normalize([n, d])

s = input()

data = [x.split("/") for x in s.split()]

before = normalize([int(x) for x in data[0]])

for i in range(1, len(data), 2):
  next = list(map(int, data[i+1]))

  if data[i][0] == '+':
    before = plus(before, next)
  elif data[i][0] == '-':
    before = minus(before, next)

print(str(before[0]) + "/" + str(before[1]))

### END SOLUTION
````

#### **10-3 題目說明：**
#### 一個矩陣A可以經過一連串的旋轉與翻轉操作後，轉換成新矩陣B。
#### 輸入一個矩陣B和一連串的操作，請算出原始的矩陣A，[b965](https://zerojudge.tw/ShowProblem?problemid=b965)

#### **輸入說明：**

*   第一行有三個正整數 R, C, M
*   接下來有 R 行(line)是 矩陣B 的 內容，每一行(line)都包含 C 個正整數，其中的第 i 行第 j 個數字代表 矩陣Bij 的值
*   在矩陣內容後的一行有 M 個整數，表示對 矩陣A 進行的操作
*   0 代表旋轉、 1 代表翻轉


#### **解題想法：**
*   這是一個有複合型的測資輸入，如果只單純閱讀數字會無法理解和組合測資
*   依照輸入說明可以得知此測資分為三個部分
    * R, C, M
    * 矩陣內元素
    * 翻轉旋轉的動作
*   在解析測資的時候，最簡單的方式就是依照提議建立所需要的變數或串列

````python
### BEGIN SOLUTION
s = input()
data = s.split()
R = int(data[0])
C = int(data[1])
M = int(data[2])

matrix = []

for i in range(R):
    s = input()
    Ci = [int(x) for x in s.split()]
    matrix.append(Ci)

data = input()
op = list(map(int, data.split()))

print(R, C, M)
print(matrix)
print(op)
### END SOLUTION
````
