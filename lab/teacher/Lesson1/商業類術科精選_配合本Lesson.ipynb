## Problem A 二數有權重的相加–(單列)

### 題目（PDF 擷取）

```text
Problem Description
輸入:二個整數𝑎和𝑏，有權重的相加，輸出: 4 * 𝑎+ 6 * 𝑏.
Input Format
輸入的一列有二個整數，二個數字0 ≤𝑎, 𝑏≤106。只有一列。
Output Format
輸出一個整數，對於所輸入的二個整數，輸出一列，依公式有權重的相加。
Sample Input 1
1 1
Sample Output 1
10
Sample Input 2
10 100
Sample Output 2
640
Hint
```

### 參考解答


````python
num1, num2 = map(int, input().split())
print(4*num1 + num2*6)
````
