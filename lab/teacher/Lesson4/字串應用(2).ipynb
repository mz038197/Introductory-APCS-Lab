### 🦾 **APCS 9-3 split() 與重組 (自我挑戰)**
#### **題目說明：**
#### 阿妮在一間出版社負責編輯的工作，為了編修文章，常常需要增加或減少標點符號，有天電腦不小心中毒，導致文章的檔案被加入了不需要的符號，阿妮自己設計一個程式，但不知道要怎麼完成, 請你幫助他，將程式完成, 讓符號移除。
*   需要保留的符號： . , !
*   不需要的符號：() - [] {} ; : <>?@#$%^&*

#### **輸入說明：**
*   每一行有一筆文字測資

#### **輸出說明：**
*   輸出排除掉不需要符號的結果
*   需要保留空格、及指定的符號

``` python
Sample Input:
he---ll[o], wo?r#l^d!
A(P)-;@C%S&
(Pra#(c)t[ice]-- ?mak(@e&s) perf--ect.

Sample Output:
hello, world!
APCS
Practice makes perfect.
```

```python
#雜訊文字 需要移除的符號
noise = '()-[]{};:<>?@#$%^&*'

#start
s = input()
words = s.split()

correct = ''

for w in __________:
    w_len = _________
    string = ""
    for i in range(_________):
        if w[i] not in _________:
            string += w[i]
            
    correct += string+" "

print(correct)
```

#### **解題想法**
*   以純文字形式處理此問題，不需要使用數值轉換
*   空格 , . ! 為必要符號，其他則否，因此可以建立一個對照表串列，利用查表的方式移除不必要的符號
*   若輸入為一串句子，則可以使用 split() 方式切開各字詞，針對每個字詞中移除不必要的符號，最後重組文字

````python
#雜訊文字 需要移除的符號
noise = '()-[]{};:<>?@#$%^&*'

#start
s = input()
words = s.split()

correct = ''

for w in words:
    w_len = len(w)
    string = ""
    for i in range(0, w_len):
        if w[i] not in noise:
            string += w[i]

    correct += string+" "

print(correct)
````

#### **ASCII進行字元與數字轉換函式**
```python
ord(c) # 將字元轉為 ASCII 值
chr(value) # 將數值轉為 ASCII 對應的文字
```

#### **題目說明：**
#### 英文名字其實隱藏著幸運值，其規則如下

*   將英文名字中的所有字元透過 ASCII 轉換為數值
*   (奇數字元的數值加總 - 偶數字元的數字加總) 取絕對值
*   最後再對 5 取餘數

#### 請撰寫一個程式，測試每個英文名的幸運值和代表意思

```
Sample Input :
Kevin
Tony
Leo
Vivian
Charlie

Sample Output :
Lucky Guy
OK
Lucky Sheep
Good Luck
Lucky Ball
```

#### **解題想法**
*   因為測試資料無法預期何時讀取結束，因此使用連續 IPO
*   由字元轉為數字需要使用 ord() 函式

````python
import sys

bank = ['Lucky Ball',
        'Lucky Sheep',
        'Lucky Guy',
        'OK',
        'Good Luck']

#start
for s in sys.stdin:
    chrlist = s
    oddValue = 0 #奇數
    evenValue = 0 #偶數

    for i in range(len(chrlist)):
        if i%2 == 0:
            oddValue += ord(chrlist[i])
        else:
            evenValue += ord(chrlist[i])

    lucky = abs(oddValue - evenValue)%5

    print(bank[lucky])
````

### 🦾**APCS 9-5 解碼器 (自我挑戰)**
#### **題目說明：**
#### 在密碼學中有一種很簡單的加密方式，就是將明文的每個字元加上某個整數 K 而得到密碼的字元

#### **輸入說明：**
*   每一行為一筆資料，每筆資料分為兩個部分
*   欲轉換的文字，可能是加密過的結果或一般字詞
*   第二個為 K 值，用來轉換文字結果， -15 <= K <= 15

#### **輸出說明：**
*   加解密後的字串結果

```
Sample Input :
HELLO 3
FUHX -5
Foj^ed 10
Tomorrow 6
`TV[\aX 13

Sample Output :
KHOOR
APCS
Python
Zusuxxu}
machine
```

#### **解題想法**
*   測資輸入的文字為單一字詞與 K 值的組合，因此需要使用 split()將資料分開
*   轉換的流程應為：
    1.   輸入字串依序轉換為 ASCII 數值
    2.   將數值加上 K 值，數值範圍有正有負
    3.   將數值結果透過 chr() 轉換回 ASCII 碼



````python
import sys, io

DATA = """\
HELLO 3
FUHX -5
Foj^ed 10
Tomorrow 6
`TV[\aX 13
"""
sys.stdin = io.StringIO(DATA)

#start
for s in sys.stdin:
    data = s.split()
    code = data[0]
    K = int(data[1])

    decode = ""
    for i in range(len(code)):
        value = ord(code[i]) + K
        decode += chr(value)

    print(decode)
````
