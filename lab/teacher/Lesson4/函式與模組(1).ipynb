### 🦾**APCS 8-2 最短距離 (自我挑戰)**
#### **題目說明：**
#### 在凹嗚村住著眾多村民，某天卻出現了一隻狼人偽裝成村民藏在家裡，每到了夜晚他會尋找獵物吃掉一定距離為內的村民，幸好有占卜師偵測出狼人的位置。
#### 請撰寫一個程式，幫村民計算是否會被狼人攻擊

#### **輸入說明：**
*   一共有兩行，第一行有三個數字。
*   分別代表村民的位置、狼人的位置、狼人的攻擊範圍。
*   第二行是每個住家與左邊住家的距離，預設第一個住家距離為 0。

#### **輸出說明：**
*   Yes 代表會被攻擊，No 代表不會被攻擊。

```
Sample Input #1 :
1 5 10
0 4 2 3 3 5 6

Sample Output #1 :
No
```
最短距離示意圖.png
```
Sample Input #2 :
3 5 15
0 4 2 3 3 5 6

Sample Output #2 :
Yes
```
```
Sample Input #3 :
6 2 5
0 4 2 3 3 5 6

Sample Output #3 :
No
```

#### **解題想法**
*   依據測試資料樣式，此題不需要使用連續 IPO
*   範例輸入有兩行，每行的測資數量不太一樣。
*   以此題測資的狀況，可以先寫成兩行 input() 讀入測資，接著使用 split() 分割每個元素
*   第一個住家沒有左側鄰居，因此距離必定為 0。
*   若要計算兩個住家之間的距離，應該要找出正確的索引值位置，相減計算。



````python
#start

s1 = input()
s2 = input()

s1 = s1.split()
s2 = s2.split()

man = int(s1[0])
wolf = int(s1[1])
atk_range = int(s1[2])

house = [ int(i) for i in s2 ]

# 正常用法
if man < wolf:
    dist = sum(house[man+1:wolf+1])
else:
    dist = sum(house[wolf+1:man+1])

# 進階用法
# dist = sum(house[min(man, wolf)+1:max(man, wolf)+1])

if dist <= atk_range:
    print("Yes")
else:
    print("No")
````

### **APCS 8-3 自訂函式**

```python
def add(a, b):
  c = a + b
  return c
```

*   參數是一個變數資料，由函式呼叫的時候外部傳入
*   冒號代表以下程式碼若有縮排，則具有相同的區域特性
*   使用 return 關鍵字將資料傳回至呼叫此程式的位置
*   函式呼叫之前需要先定義好函式，否則程式會發生錯誤

#### **題目說明：**
#### 使用自訂函式語法 def，建立一個數值相加與文字相加的函式，並進行正確的計算

#### **輸入說明：**
*   每一行有三個資料。
*   第一項代表數值加法或字串加法。
*   後兩項為相加的資料。

```
Sample Input #1 :
1 3 5

Sample Output #1 :
8
```
```
Sample Input #2 :
2 a b

Sample Output #2 :
ab
```
```
Sample Input #3 :
1 10 12

Sample Output #3 :
22
```
```
Sample Input #4 :
2 P A

Sample Output #4 :
PA
```
```
Sample Input #5 :
2 5 G

Sample Output #5 :
5G
```

#### **解題想法**
*   此題需要使用連續 IPO。
*   每組測資第一個數字 = 1，代表數值相加計算；如果第一個數字 = 2，代表字串相加運算。
*   可以使用條件區分要進行的運算，並分成兩個函式。
*   也可以先建立一個加法函式，再使用條件式區分計算。

````python
import sys

def plus(a, b):
  return int(a) + int(b)

def combine(a, b):
  return str(a) + str(b)

#start
while True:
    s = input()
    inputs = s.split()
    mode = int(inputs[0])
    if mode == 1:
        print(plus(inputs[1], inputs[2]))
    elif mode == 2:
        print(combine(inputs[1], inputs[2]))
````

### 🦾**APCS 8-4 溫度轉換 (自我挑戰)**
#### **題目說明：**
#### 溫度有兩種單位，分別是攝氏和華氏，但這兩種溫度當中存在一個轉換的公式，可以讓兩種溫度找到一個對應關係。
#### 輸入為攝氏溫度，輸出為華氏溫度

#### 轉換公式：華氏 = 攝氏*(9/5) + 32

```
Sample Input :
0
15
-5
30

Sample Output :
32
59
23
86
```



#### **解題想法**

### 🦾**APCS 8-5 四則運算 (自我挑戰)**
#### **題目說明：**
#### 計算機是生活中的好幫手，接下來的程式需要寫一個簡單型計算機，輸入兩個數字和加減乘除符號後輸出對應的結果

#### 計算機的規則如下：
*   由於是簡易計算機，只要進行兩個數字的運算
*   此程式不需要判斷括號
*   四則運算符號如: +, -, *, /


#### **輸入說明：**
*   每一行則代表一個運算
*   數字和運算符號中間都有空白隔開

#### **輸出說明：**
*   顯示運算的結果

```
Sample Input :
4 + 5
2 - 8
12 / 3
7 * 9
13 / 0

Sample Output :
9
-6
4.0
63
N/A
```

#### **解題想法**
*   此題需要使用連續 IPO
*   每一行數字和運算符號都有空白隔開，直接使用 split() 輸入處理
*   此題需要撰寫四個函式 add(), sub(), mul(), div() 分別表示加、減、乘、除
*   最後一個測資出現 N/A 代表程式需要做除以 0 的判斷，避免程式出現錯誤訊息

````python
import sys

def add(a, b):
  return a + b

def sub(a, b):
  return a - b

def mul(a, b):
  return a * b

def div(a, b):
  return a / b

#start
for s in sys.stdin:

  inputs = s.split()
  a = int(inputs[0])
  b = int(inputs[2])
  op = inputs[1]

  if op == "+":
    print(add(a, b))
  elif op == "-":
    print(sub(a, b))
  elif op == "*":
    print(mul(a, b))
  elif op == "/":
    if b == 0:
      print("N/A")
    else:
      print(div(a, b))
````

### **APCS 8-8 認識值域**
#### **題目說明：**
*   程式是一個常出現於識讀題中的題目, 請問輸出結果為何？
*   如果想要在 calc() 中改變外部的 x，讓變數相通，該如何修改程式？

```python
# 全域變數
x = 5
y = 10

def calc():
  # 區域變數
  x = 7
  z = 12
  print( x + z )

#start
print(x)
calc()
print(x, y)
```




#### **解題想法：**
*   必須使用 global
*   在 calc() 中將 x 規範為 global
*   就能調整為全域變數 x

````python
# 全域變數
x = 5
y = 10

def calc():
  global x
  x = 7
  z = 12
  print( x + z )

# start
print(x)
calc()
print(x, y)
````

#### **注意事項**
*   **區域變數**：定義在函式中的變數，有效範圍在該函式中
*   **全域變數**：定義在函式外的變數，有效範圍為整個 .py 檔
*   當區域變數名稱與全域變數相同時，則區域變數會「隱藏」全域變數的效力，在函式中以區域變數為主
*   在函式中宣告某個區域變數為 global，則會將全域變數的效力代入該區域變數中，並且在函式中亦能修改全域變數值
*   函式傳遞參數的方式會比使用 global 安全
