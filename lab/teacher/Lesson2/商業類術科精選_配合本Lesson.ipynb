## Problem B 剪刀石頭布單行版

### 題目（PDF 擷取）

```text
Problem Description
剪刀石頭布單行版
Input Format
玩家1出的拳：[Y]剪刀[O]石頭[X]布和玩家2出的拳：[Y]剪刀[O]石頭[X]布。
Sample Input ”Y,O,X”為英文字母。　
Output Format
平手:0，玩家1贏了:1，玩家2贏了:2。輸出為0,1,2。
Sample Output ”0,1,2”為數字。　
Sample Input 1
Y Y
Sample Output 1
0
Sample Input 2
Y O
Sample Output 2
2
Sample Input 3
Y X
Sample Output 3
1
Sample Input 4
O Y
Sample Output 4
1
Sample Input 5
O O
Sample Output 5
0
Sample Input 6
O X
Sample Output 6
2
Sample Input 7
X Y
Sample Output 7
2
Sample Input 8
X X
Sample Output 8
0

Sample Input 9
X O
Sample Output 9
1
Hint
```

### 參考解答


````python
player1, player2 = input().split()


if player1 == player2:
  result = "0"
else:
  if (player1 == "Y" and player2 == "X")  or (player1 == "O" and player2 == "Y")  or (player1 == "X" and player2 == "O"):
    result = "1"
  else:
    result = "2"

print(result)
````

## Problem C 閏年

```text
西元年份(2008)，輸入民國年份(110)換成西元年份(2021)，輸入民國年份(-11)換成西元年

再算西元年份是否為閏年(True)或平年(False)。
```

## Problem D ASCII碼

### 題目（PDF 擷取）

```text
Problem Description
在電腦內部，所有的資訊都被視為數字。例如，在ASCII碼中，大寫字母’A’被賦予數
值65。同樣地，大寫字母’B’被賦予66，’C’被賦予67，等等。以此方式，’A’到’Z’依序被賦
予65至90。在ASCII碼中，小寫字母’a’被賦予97。同樣地，小寫字母’b’被賦予98，’c’被賦
予99，等等。以此方式，’a’到’z’依序被賦予97至122。
寫一個程式，以確定在ASCII碼內給定的數字是代表大寫(uppercase)還是小寫(lowercase)的
字母。
Input Format
每組輸入只有一列，內含有一個整數𝑁，0 ≤𝑁≤127。
Output Format
對於給定的數字，
大寫(uppercase)字母’A’到’Z’，’A’到’Z’依序被賦予65至90，輸出1；
小寫(lowercase)字母’a’到’z’，’a’到’z’依序被賦予97至122，輸出2；
其他字元輸出0。
Sample Input 1
67
Sample Output 1
1
Sample Input 2
97
Sample Output 2
2
Sample Input 3
31
Sample Output 3
0
Hint
```

### 參考解答


````python
num = int(input())
if (num >= ord("A") and num <= ord("Z")):
  print("1")
elif (num >= ord("a") and num <= ord("z")):
  print("2")
else:
  print("0")
````

## Problem E 2的冪

### 題目（PDF 擷取）

```text
Problem Description
將給定的數字轉換為小於或等於該數字的最大2的冪。例如，如果數字是2或3，將其轉換
為21 = 2。同樣，如果數字是4、5、6或7，將其轉換為22 = 4；如果數字是8、9、10、11,...,15，
將其轉換為23 = 8。
Input Format
每組輸入只有一列，內含有一個整數𝑁，2 ≤𝑁≤106。
Output Format
對於給定的數字，輸出轉換為小於或等於該數字的最大2的冪。

8

52

32
```

### 參考解答


````python
N = int(input())

num = 2
while num <= N:
  num *= 2

print(num // 2)
````

## Problem H 數字運算(減1或除以10)

### 題目（PDF 擷取）

```text
Problem Description
數字運算(減1或除以10)
如果數字的最後1位不是零，就將數字減1；如果數字的最後1位是零，就將數字除以10（即
去掉最後一位）。
給定一個整數𝑛。對這個數字進行𝑘次數字運算。你的任務是在進行𝑘次減1或除以10的運算
後，輸出最終結果。保證最終結果會是正整數。正整數大於0的整數。
Input Format
每組輸入只有一列，包含兩個整數𝑛和𝑘, 2 ≤𝑛≤109, 1 ≤𝑘≤50）分別是要數字運算和
減1或除以10的次數。
Output Format
輸出一個整數，將𝑛減1或除以10 𝑘次的結果。保證結果將是正整數。
Sample Input 1
512 4
Sample Output 1
50
Sample Input 2
1000000000 9
Sample Output 2
1
Sample Input 3
1001 2
Sample Output 3
100
Hint
解題思路：
在Sample Input 1中，對應於以下序列：512 →511 →510 →51 →50。
```

### 參考解答


````python
N, K = map(int, input().split())

def recursive(last, curr, times):

  if times == K:
    return curr

  if curr <= 0:
    return last

  times += 1
  last = curr

  if str(curr)[-1] != "0":
    curr -= 1
  else:
    curr //= 10

  return recursive(last, curr, times)

print(recursive(N, N, 0))
````

## Problem B 偶數

### 題目（PDF 擷取）

```text
Problem Description
請設計一個程式，輸入兩個整數，代表一個範圍的起始值和結束值(含邊界），計算這個範
圍(含邊界)中所有偶數的總和。
例如：兩個整數1 10，範圍1 到10 中的偶數有2, 4, 6, 8, 10，總和為30。
Input Format
輸入只有一行，其中含有兩個整數𝑎, 𝑏(0 ≤𝑎, 𝑏≤2147483647)（以空格隔開），輸入的整
數𝑎, 𝑏，可能為𝑎≤𝑏或𝑎≥𝑏。
Output Format
請輸出一個整數，代表𝑎與𝑏之間(含𝑎與𝑏) 所有偶數的和(答案會≤2147483647)。
Sample Input 1
2 5
Sample Output 1
6
Sample Input 2
9 7
Sample Output 2
8
Sample Input 3
1 10
Sample Output 3
30
```

### 參考解答


````python
nums = list(map(int, input().split()))
start, end = min(nums), max(nums)
result = []

for i in range(start, end+1):
  if i % 2 == 0:
    result.append(i)

print(sum(result))
````

## Problem C 三角形

### 題目（PDF 擷取）

```text
Problem Description
在數學課的三角形單元中，老師設計了一個實作作業：
1. 每位學生需要在星期一到星期五，每天攜帶三根樹枝到校。
2. 若當天帶來的三根樹枝可以構成三角形，則獲得1 分。
3. 判斷三角形的條件：任意兩邊之和必須大於第三邊。
給定某學生這五天帶來的樹枝長度資料，請計算某學生最後能獲得幾分？
Input Format
輸入一共有5 行，每行有3 個正整數𝑥𝑖，正整數1 ≤𝑥𝑖≤30 ，代表某學生當天所帶的樹枝的
長度，每兩個正整數之間用一個空格隔開。
Output Format
輸出一個整數，代表某學生可以獲得幾分。
Sample Input 1
1 2 3
2 4 6
3 4 5
5 3 2
1 3 5
Sample Output 1
1
Sample Input 2
18 6 4
12 10 17
12 21 21
16 6 28
18 26 20
Sample Output 2
3
```

### 參考解答


````python
total = 0
for _ in range(5):
  s = input()
  a, b, c = list(map(int, s.split()))
  if a + b > c and b + c > a and a + c > b:
    total += 1

print(total)
````

## Problem K 四捨五入

### 題目（PDF 擷取）

```text
Problem Description
在日常生活中，小數點後的細微差異往往意義不大。以氣溫為例，攝氏15度和15.05度對人體
的感受幾乎沒有區別。為了簡化數據處理，我們可以使用四捨五入的方法將數字化整，使其
更容易理解與統計。
請設計一個程式，將輸入的小數進行四捨五入處理。
Input Format
第一行有一個整數𝑛(0 < 𝑛≤10000)，表示有幾個測試資料。
接下來的𝑛行，每行有一個小數𝑥(−1 ≤𝑥≤1)。
Output Format
請輸出四捨五入至小數點以下第二位的結果，如果輸出結果-0.00，因為零沒有正負號，所以
輸出應修正為0.00。
Sample Input 1
10
1.00000
0.5
0.715
0.1234567890
-0.995
0.315
0.316
0.415
0.416
-0.00325
Sample Output 1
1.00
0.50
0.72
0.12
-1.00
0.32
0.32
0.42
0.42
0.00
11
```

### 參考解答


````python
times = int(input())
EPSILON = 1e-10

for _ in range(times):
  num = float(input())
  num += (-1*EPSILON) if num < 0 else EPSILON
  num = round(num, 2)
  if num == 0:
    print("0.00")
  else:
    print(f"{num:0.2f}")
````
