## **基礎迴圈** (100 min)
### **重複執行的意思**

### **基礎迴圈有兩種**
*   while 迴圈 : 條件判斷為主要控制
*   for 迴圈 : 計數迴圈

### **while 迴圈需要注意的事情：**


*   判斷是否重複執行完全依靠條件決定，如果邏輯設計有誤，則可能發生無限迴圈的情況
*   可以使用 break, continue 關鍵字中斷迴圈的執行



<br>

### **APCS 5-1 迴圈的行為**
#### **題目說明：**
#### 練習使用 while 迴圈，連續輸出 1~5 的數字
``` python
while 需要滿足的條件:
  # 回圈內要執行的程式
  #...
  #...
```

#### **解題想法**
*   while 迴圈是一個需要條件的迴圈，在這個題目中，什麼時候需要離開結束迴圈呢？
*   輸出數字 1 ~ 5 代表程式一開始會有一個數值為 1 ，接著依序累加到 5



````python
#start

num = 1

while num<=5 :
    print(num)
````

#### **緊急!**
*   發生"無限迴圈"!!!
*   程式停不下來的時候可以按下鍵盤的 "ctrl+c"

````python
#start

num = 1

while num<=5 :
    print(num)
    num = num+1
````

### **APCS 5-4 找出因數**
#### **題目說明：**
#### 輸入一個正整數 N，列出此數字的所有因數

#### **範例輸入：**
*   10
*   21
*   48

#### **應得輸出：**
*   1 2 5 10
*   1 3 7 21
*   1 2 3 4 6 8 12 16 24 48


#### **解題想法**
#### 此題需要注意的重點
*   目前學會兩種迴圈的語法，依照此題的狀況，你會選擇哪種？
    1. 無法預測或估算迴圈次數，使用 while 迴圈 <br>
    2. 能估算或具有特定規律性，使用 for 迴圈 <br>
*   數學中，如何判斷因數？
*   此題不需要寫成連續 IPO 結構

````python
#start

N = int(input())
i=1
while i<=N:
    if N%i == 0:
        print(i, end=' ')
    i=i+1
````

````python
#start
N = int(input())
for i in range(1, N+1):
  if N % i == 0:
    print(i, end=" ")
````

### **APCS 5-5 數列求和 (自我挑戰)**
#### **題目說明：**
#### 輸入一個正整數 N，請輸出 1+2+3+ … +N 的總和





#### **解題想法：**
*   由題目判斷使用 while or for 迴圈
*   求總和必須增加一個變數，用來儲存目前累加的結果

````python
#start

N = int(input())

sum = 0

for i in range(1,N+1):
    sum = sum+i

print(sum)
````

### **APCS 5-6 數列階乘 (自我挑戰)**
#### **題目說明：**
#### 數學中有一個特殊的用法稱為階乘 (factorial)
#### 輸入一個正整數 N，輸出 1x2x3x … xN 的結果


#### **範例輸入：**
*   4
*   24

#### **應得輸出：**
*   12
*   479001600



#### **解題想法：**
*   由題目判斷使用 while or for 迴圈
*   求階乘的結果需要不斷的連續相乘數字



````python
#start

N = int(input())

factorial = 1

for i in range(1,N+1):
    factorial = factorial*i

print(factorial)
````

### **APCS 5-7 99乘法表**
#### **題目說明：**
#### 請利用 for 迴圈建立一個 99 乘法表，輸出請參考以下結果
```python
1x1= 1 1x2= 2 1x3= 3 ...
2x1= 2 2x2= 4 2x3= 6 ...
3x1= 3 3x2= 6 3x3= 9 ...
...
...
9x1= 9 9x2= 18 9x3= 27 ...
```

#### **解題想法：**
*   只用一次 for 迴圈好像沒辦法建立 1x1 ~ 9x9
*   輸出不是只有乘法結果，連算式都需要輸出

````python
#start

for i in range(1, 10):
    for j in range(1, 10):
        if i*j <10:
            print(str(i)+"x"+str(j)+"= "+str(i*j), end=' ')
        else:
            print(str(i)+"x"+str(j)+"="+str(i*j), end=' ')
    print()
````

### **APCS 5-8a 文字圖形產生器**
#### **題目說明：**
#### 用迴圈與條件式用在排版輸出的應用
#### 透過輸入 M, N 兩個值，可以控制長方形的形狀
```python
*****
*****
*****
*****
```

````python
#start

M = int(input())
N = int(input())

for i in range(N):
    for j in range(M):
        print("*", end='')
    print()
````


````python
#start

M = int(input())
N = int(input())

for i in range(5+1):
    for j in range(1, i+1):
        print("*", end='')
    print()

for i in range(1,6):
    print(" "*(5-i)+"*"*i)
````

### **APCS 5-9 文字圖形產生器(進階)**
#### **題目說明：**
#### 請使用迴圈加條件式產生以下文字圖形

#### APCS 5-9a.py
```
    *
   * *
  * * *
 * * * *
* * * * *
 * * * *
  * * *
   * *
    *
```

````python
#start

for i in range(-4, 4+1):
    print(abs(i)*" ", end='')
    for j in range(1, 5-abs(i)+1):
        print(" *", end='')
    print()
````

## 重點複習

*   迴圈是條件式以外的另外一種流程控制
*   迴圈有兩種類型 while 迴圈，for 迴圈
*   while 迴圈主要用在不知道會執行幾次的程式
*   for 迴圈通常會搭配 range() 函數，執行固定的重複次數
*   迴圈除了能重複計算以外，也能搭配條件式使用
*   雙層迴圈是進階的用法，可以重複多層次的迴圈結構
