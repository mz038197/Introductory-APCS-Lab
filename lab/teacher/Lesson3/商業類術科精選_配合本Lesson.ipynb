## Problem G 因數加總

### 題目（PDF 擷取）

```text
Problem Description
數字20 的因數包括所有能夠整除20 的數字。這些因數是：
1, 2, 4, 5, 10, 20
這些數字可以成對地乘起來得到20，例如1 × 20、2 × 10、4 × 5 等等。
數字30 的因數是那些能夠整除30 的數字。這些因數包括：
1, 2, 3, 5, 6, 10, 15, 30
這些數字成對相乘會得到30，例如1 × 30、2 × 15、3 × 10、5 × 6 等等。
請找出輸入數字的所有因數，並且把所有因數加總後並輸出
例如:
輸入數字為20，因數為1,2,4,5,10,20，相加為42
輸入數字為30，因數為1,2,3,5,6,10,15,30，相加為72
輸入數字為999997，因數為1,757,1321,999997，相加為1002076
Input Format
每組輸入只有一列，包含一個整數𝑁，2 ≤𝑁≤106
Output Format
針對輸入整數，把所有因數加總後並輸出。
Sample Input 1
20
Sample Output 1
42
Sample Input 2
30
Sample Output 2
72
Sample Input 3
999997
Sample Output 3
1002076
Hint
```

### 參考解答


````python
N = int(input())

c = []
for i in range(1, N+1):
  if N%i == 0:
    c.append(i)

print(sum(c))
````

````python
times = int(input())


for _ in range(times):
  n = int(input())
  nums = list(map(int, input().split()))
  total = nums.pop()

  result = {}
  is_matched = False
  for num in nums:
    if total - num in result:
      is_matched = True
      break
    else:
      result[num] = 1

  if is_matched:
    print("Yes")
  else:
    print("No")

````

### 題目（PDF 擷取）

```text
Problem Description
在統計學的世界中，中位數（median）扮演一個重要的角色。根據定義：中位數就是在一連
串已由小到大排序的數字中，排在中間的那一個數。
例如：在1,3,4,9,11中，4就是中位數。萬一有偶數個數字，我們定義中位數就是位於中間的
那2個數的和除以2（而且只取整數部分）。
例如：在1,2,3,6,7,8中3,6是位於中間的2個數，所以中位數就是(3+6)/2=4。
Input Format
輸入資料的第一列包含一個整數𝑡(1 ≤𝑡≤103)為測試案例的數量。輸入包含了𝑡個整

7

1
2
3

4
Hint
數值一直由小到大插入陣列中，即可得其中位數。
```

### 參考解答


````python
t = int(input())
arr = []
for _ in range(t):
  arr.append(int(input()))
  arr.sort()
  mid = len(arr) // 2
  if len(arr) % 2 == 0:
    mid = len(arr) // 2
    print((arr[mid-1] + arr[mid])//2)
  else:
    print(arr[mid])
````

## Problem O 洗牌

### 題目（PDF 擷取）

```text
Problem Description
撲克牌是非常適合休閒娛樂的卡牌遊戲，依據不同規則可以衍生多種遊戲，例如二十一點和
```

### 參考解答


````python
N, M = map(int, input().split())

cards = list(map(int, input().split()))

for _ in range(M):
  result = []
  mid = len(cards) // 2
  part1 = cards[:mid]
  part2 = cards[mid:]
  for i in range(mid):
    result.append(part1[i])
    result.append(part2[i])
  cards = result

print(*cards)
````

## Problem D 不同的數字

### 題目（PDF 擷取）

```text
Problem Description
給一個整數陣列，你的任務是計算這個陣列之中相異整數的數量。
Input Format
第一行有一個整數𝑛（1 ≤𝑛≤105）。
第二行有𝑛個整數𝑥𝑖（1 ≤𝑥𝑖≤106），每兩個正整數之間用一個空格隔開。
Output Format
輸出相異整數的數量。
Sample Input 1
5
2 3 2 2 3
Sample Output 1
2
Sample Input 2
6
1 1 2 2 3 4
Sample Output 2
4
```

### 參考解答


````python
times = int(input())
nums = list(map(int, input().split()))

result = {}
for i in range(times):
  if nums[i] not in result:
    result[nums[i]] = 1
  else:
    result[nums[i]] += 1

print(len(result))
````

## Problem E 眾數

### 題目（PDF 擷取）

```text
Problem Description
給定多個無序正整數，同一個無序正整數可能會出現多次，出現次數最多的整數稱為眾數。
求出它的眾數及它出現的次數。
Input Format
給定一個整數𝑛(1 ≤𝑛≤105)，接下來的一行有𝑛個整數𝑥𝑖(1 ≤𝑥𝑖≤106)無序正整數，每兩
個正整數之間用一個空格隔開。
Output Format
輸出結果，若有多行（因為眾數可能不唯一）每行兩個數，結果眾數需依照字典序從小到大
排列。第1個是眾數，第2個是眾數出現的次數（以空格隔開）。
Sample Input 1
12
2 4 2 3 2 5 3 7 2 3 4 3
Sample Output 1
2 4
3 4
Sample Input 2
12
1 1 1 3 3 4 4 3 4 2 2 2
Sample Output 2
1 3
2 3
3 3
4 3
```

### 參考解答


````python
times = int(input())
nums = list(map(int, input().split()))

result = {}
for i in range(times):
  if nums[i] not in result:
    result[nums[i]] = 1
  else:
    result[nums[i]] += 1

maxTimes = max(result.values())
for key in sorted(list(result.keys())):
  if result[key] == maxTimes:
    print(f"{key} {maxTimes}")
````

````python
m, n = list(map(int, input().split()))

result = [] # 2d
for _ in range(m):
  row = list(map(int, input().split()))

  #將row翻轉存進result
  for i in range(n):
    if len(result) <= i:
      result.append([])
    result[i].append(row[i])

for row in result:
  print(*row)
````
