### **APCS 6-3 串列與連續 IPO**
#### **題目說明：**
#### 輸入 N 個數字，請將 N 個數字轉存為整數串列，並輸出平均值
#### 此題需要使用連續 IPO

#### **範例輸入:**
*   1 2 3 4 5
*   5 9 12 6 8 10 0 3
*   9 100 20 5

#### **應得輸出：**
*   3.0
*   6.625
*   33.5


#### **解題想法**
*   此題需要使用連續 IPO。
*   split() 分割出來的元素為字串型態，因此需要將元素轉成整數
*   如果數字越多，該如何快速又簡潔的將所有元素型態轉換？
*   輸入資料並沒有說明 N 為多少，如何知道串列長度？

````python
import sys

#start
for s in sys.stdin:
    nums = s.split()

    size = len(nums)

    for i in range(size):
      nums[i] = int(nums[i])

    print(sum(nums)/size)
````

#### **細節注意!**

*   Python 中 split() 方法在沒有指定參數時，預設會使用空白字元（whitespace）作為分隔符號
*   split() 在沒有參數時還有一個特殊行為：它會自動移除字串開頭和結尾的空白字元，並且連續的空白字元會被視為單一分隔符號


#### 健康檢查的時候，需要知道以下幾種狀況

*   計算目前隊伍的總人數
*   增加一個學生進入排隊隊伍最尾端
*   從隊伍中移除一個不遵守秩序的學生
*   將目前隊伍中的學生依照號碼重新排序

*   每一行可能會有 1~2個數字
*   第一個數字代表進行的動作
*   第二個數字代表學生的號碼 (部分操作可能沒有第二個數字)

#### **動作編號對應**
*   1 --> 計算總人數
*   2 --> 加入一個學生至串列尾端

*   4 --> 排序學生的號碼

*   此題需要使用連續 IPO
*   第一次碰到輸入的資料會有浮動的狀態，處理輸入資料時可能會需要知道每一筆輸入資料的長度

````python
import sys

#start
student = [4, 3, 2, 1]

for s in sys.stdin:
    op = s.split()

    if op[0] == "1":
        print(len(student))
    elif op[0] == "2":
        student.append(int(op[1]))
        print(student)
    elif op[0] == "3":
        student.remove(int(op[1]))
        print(student)
    elif op[0] == "4":
        student.sort()
        print(student)
````

#### **字典結構使用時的注意事項：**
*   字典無法使用索引值的方式存取，只能使用 key 查詢 value
*   字典不能排序，元素之間也沒有順序關係
*   字典結構常用在需要有對應關係的兩組資料中，若程式有這樣的特性則可以使用字典
*   建立字典必須使用 {}，但以鍵(key)存取值(value) 則是使用 []

#### **用串列模擬字典實作 (延伸學習)**

````python
class Hashtable:
	def __init__(self):
		"""
		Create an array(self.mydict) with a bucket size - which is derived from the load factor.
		The Load factor is a measure that decides when to increase the HashMap capacity to maintain the get() and put() operation complexity of O(1).
		The default load factor of HashMap is 0.75f (75% of the map size).
		Load Factor = (n/k)
		where n is the number of max number of elements that can be stored dict
		k is the bucket size
		Optimal Load factor is around (2/3) such that the effect of hash collisions is minimum
		"""
		self.bucket = 16
		self.hashmap = [[] for i in range(self.bucket)]

	def __str__(self):
		return str(self.__dict__)

	def hash(self, key):
		return len(key) % self.bucket

	def put(self, key, value):
		"""
		value may already be present
		"""
		hash_value = self.hash(key)
		reference = self.hashmap[hash_value]
		for i in range(len(reference)):
			if reference[i][0] == key:
				reference[i][1] = value
				return None
		reference.append([key, value])
		return None

	def get(self, key):
		"""
		Returns the value to which the specified key is mapped, or -1 if this map contains no mapping for the key
		"""
		hash_value = self.hash(key)
		reference = self.hashmap[hash_value]
		for i in range(len(reference)):
			if reference[i][0] == key:
				return reference[i][1]
		return -1

	def remove(self, key):
		"""
		Removes the mapping of the specified value key if this map contains a mapping for the key
		"""
		hash_value = self.hash(key)
		reference = self.hashmap[hash_value]
		for i in range(len(reference)):
			if reference[i][0] == key:
				reference.pop(i)
				return None
		return None

h=Hashtable()
h.put('grapes',1000)
h.put('apples',10)
h.put('ora',300)
h.put('banan',200)
print(h.get('grapes'))
print(h)
h.remove('apples')
print(h)
````
