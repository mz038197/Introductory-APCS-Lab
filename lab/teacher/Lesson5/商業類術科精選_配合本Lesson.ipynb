## Problem J 3n+1

### 題目（PDF 擷取）

```text
Problem Description
有個演算法接受一個正整數𝑛，並重複以下步驟直到𝑛變成1：
- 如果𝑛是偶數，就將它除以2；
- 如果𝑛是奇數，就將它乘以3 再加1。
例如從𝑛= 3 開始，演算法的過程：
𝑛= 3 是奇數，所以讓𝑛= 3 × 3 + 1 = 10；𝑛= 10 是偶數，所以讓𝑛= 10 ÷ 2 = 5；𝑛= 5 是
奇數，所以讓𝑛= 5 × 3 + 1 = 16；以此類推，這個過程會是
[︀
3 →10 →5 →16 →8 →4 →2 →1
]︀
給你一個正整數𝑛，你的任務是模擬這個演算法。
Input Format
一個正整數𝑛。（1 ≤𝑛≤106）
Output Format
將正整數𝑛經過演算法產生的所有數值按照順序輸出成一行。
也就是輸出𝑛𝑘1 𝑘2 · · · 1，其中演算法的過程為𝑛→𝑘1 →𝑘2 →· · · →1（以空格隔開）。
Sample Input 1
3
Sample Output 1
3 10 5 16 8 4 2 1
```

### 參考解答


````python
def calculate(num):
  print(num, end=" ")
  if num == 1:
    return

  if num % 2 == 0:
    calculate(num//2)
  else:
    calculate(num * 3 + 1)

num = int(input())

calculate(num)
````
