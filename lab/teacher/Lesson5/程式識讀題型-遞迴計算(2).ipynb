#### **1-1 代值計算：**
*   傳入指定參數，詢問回傳值
*   逐步推算即可得到計算結果

#### **例題**
#### 給定下面函式 g()，請問 g(13) 的回傳值為何？
```python
def g(a):
    if a<=1:
        return a
    else:
        return g(a-2)+3
```
*   (A) 16
*   (B) 18
*   (C) 19
*   (D) 22




#### **解析過程**
```python
g(13) = g(11) + 3
g(11) = g(9) + 3
g(9) = g(7) + 3
g(7) = g(5) + 3
g(5) = g(3) + 3
g(3) = g(1) + 3
g(1) = 1
```
#### **經歷幾次**
``` python
g(1) = 1
g(n) = g(n-2) + 3
g(13) = 首項 + ((起始 - 終點) / 間隔 + 1) * 公差
      = 1 + ((11 - 1) / 間格 + 1) * 3
      = 19 (C)
```

#### **1-2 追蹤遞迴過程的行為：**
*   了解遞迴函式執行過程中，變數、計算數值或輸出的變化

#### **例題**
#### 請問 func(123) 的輸出結果為何？
```python
def func(n):
    if n == 0:
        return 0
    print(n, end=" ")
    tmp = n % 10 + func(n//10)
    print(tmp, end=" ")
    return tmp
```
*   (A) 123 15 6 1 0
*   (B) 0 1 6 15 123
*   (C) 123 12 1 3 6 123
*   (D) 123 12 1 1 3 6


#### **解析過程**
```python
func(123) = func(12) + 123 % 10
func(12) = func(1) + 12 % 10
func(1) = func(0) + 1 % 10
func(0) = 0
```
#### **經歷幾次**
``` python
func(123) = 123 % 10 +  12 % 10 + 1 % 10 + 0
          = 3 + 2 + 1
          = 6 (D)
```

#### **例題2**
#### 下面函式兩個回傳式，分別該如何撰寫，才能正確計算，並回傳 a, b 之間的最大公因數（GCD）？

```python
def GCD(a, b):
    r = a % b
    if r == 0:
        return ___(I)___
    return ___(II)___
```
*   (A) I: a, II: GCD(b, r)
*   **(B) I: b, II: GCD(b, r)**
*   (C) I: a, II: GCD(a, r)
*   (D) I: b, II: GCD(a, r)

#### **演進過程**
```
#1, a=15, b=12
    GCD(15, 12)
    r = 15%12 = 3

#2, a=12, b=3
    GCD(12, 3)
    r = 12%3 = 0
    return 3 (3為最大公因數)

=> GCD(a, b) = GCD(b, a mod b)
```


````python
def GCD(a, b):
    r = a % b
    if r == 0:
        return b
    return GCD(b, r)

print(GCD(15, 12))
````

````python
# 迴圈版本
def GCD(a, b):
  while a != b:
    if a > b:
      a = a - b
    else:
      b = b - a
  return a

print(GCD(15, 12))
````
