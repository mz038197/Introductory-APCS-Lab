*   在堆疊頂端加入元素。

*   刪除堆疊最頂端的元素


````python
# 使用 串列(List) 實作
#start
class Stack:

  def __init__(self):
    self.stack = []

  def push(self, data):
    """ 將資料加入堆疊 """
    self.stack.append(data)

  def pop(self):
    """ 將頂端資料從堆疊中移出 """
    if not self.empty():
      return self.stack.pop()

  def empty(self):
    """ 檢查堆疊是否為空 """
    return len(self.stack) == 0

  def top(self):
    """ 顯示目前在堆疊頂端的資料 """
    if not self.empty():
      return self.stack[-1]

  def size(self):
    """ 顯示目前堆疊中的資料數量 """
    return len(self.stack)

s = Stack()
s.push(1)
s.push(2)
s.push(3)
print(s.pop())
print(s.pop())
print(s.pop())
````

*   在隊伍尾端加入元素。
*   輸出隊伍最前端的元素。
*   刪除隊伍最前端的元素。

````python
# 使用 串列(List) 實作
class Queue:

  def __init__(self, MAX_SIZE=10):
    self._queue = []
    self._max_size = MAX_SIZE

  def enqueue(self, data):
    if len(self._queue) >= self._max_size:
      return False
    else:
      self._queue.append()
      return True

  def dequeue(self):
    if len(self._queue) == 0:
      return False
    else:
      self._queue.pop(0)
      return True

  def isEmpty(self):
    return len(self._queue) == 0

  def isFull(self):
    return len(self._queue) == self._max_size

  def front(self):
    if self.isEmpty():
      return None
    else:
      return self._queue[0]

  def rear(self):
    if self.isEmpty():
      return None
    else:
      return self._queue[-1]
````

#### 🦾 **[232. Implement Queue using Stacks](https://leetcode.com/problems/implement-queue-using-stacks/description/) (終極挑戰)**
#### **題目說明：**
#### 僅使用兩個堆疊實作先進先出（FIFO）佇列。所實作的佇列應支援一般佇列的所有功能（ push 、 peek 、 pop 和 empty ）。

#### 實作 MyQueue 類別：
*  void push(int x) 將元素 x 推到佇列的尾端。
*  int pop() 移除佇列前端的元素並回傳它。
*  int peek() 回傳佇列前端的元素。
*  boolean empty() 若佇列為空則回傳 true ，否則回傳 false 。

#### 注意事項：

*  你必須只使用堆疊的標準操作，這表示只有 push to top 、 peek/pop from top 、 size 與 is empty 這些操作是有效的。
*  依據你使用的程式語言，堆疊可能不是原生支援。你可以使用串列或 deque（雙端佇列）來模擬堆疊，只要你僅使用堆疊的標準操作即可。

```
Sample Input :
["MyQueue", "push", "push", "peek", "pop", "empty"]
[[], [1], [2], [], [], []]

Sample Output :
[null, null, null, 1, 1, false]
```

````python
class MyQueue:
    def __init__(self):
        self.push_stack = []
        self.pop_stack = []
        self.front = None

    def push(self, x: int):
        if not self.push_stack:
            self.front = x
        self.push_stack.append(x)

    def pop(self) -> int:
        if self.pop_stack:
            return self.pop_stack.pop()
        while self.push_stack:
            self.pop_stack.append(self.push_stack.pop())
        return self.pop_stack.pop()

    def peek(self) -> int:
        if self.pop_stack:
            return self.pop_stack[-1]
        return self.front

    def empty(self) -> bool:
        return not self.push_stack and not self.pop_stack
````

````python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self,data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
            self.tail = self.head
            self.length = 1
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node
            self.length += 1

    def prepend(self,data):
        new_node = Node(data)
        new_node.next = self.head
        self.head.prev = new_node
        self.head = new_node
        self.length += 1

    def insert(self,index,data):
        new_node = Node(data)
        if index==0:
            self.prepend(data)
            return
        if index >= self.length:
            self.append(data)
            return
        else:
            leader = self.traversetoindex(index - 1)
            holder = leader.next
            leader.next = new_node
            new_node.next = holder
            new_node.prev = leader
            holder.prev = new_node
            self.length+=1

    def remove(self,index):
        if index==0:
            self.head=self.head.next
            self.length-=1
            return
        if index == self.length-1:
            self.tail = self.tail.prev
            self.tail.next = None
            self.length -= 1
            return
        leader = self.traversetoindex(index-1)
        unwanted_node = leader.next
        holder = unwanted_node.next
        leader.next = holder
        holder.prev = leader
        self.length -= 1


    def traversetoindex(self,index):
        curr_node = self.head
        i = 0
        while i!= index:
            curr_node = curr_node.next
            i+=1
        return curr_node

    def printt(self):
        temp = self.head
        while temp != None:
            print(temp.data , end = ' ')
            temp = temp.next
        print()
        print('Length ' + str(self.length))


d = DoublyLinkedList()
d.append(10)
d.append(5)
d.append(6)
d.prepend(1)
d.insert(2,22)
d.remove(3)
d.printt()
````
